### 3. Utilizzo delle API

Le API di TMDb forniscono una vasta gamma di funzionalità, inclusi:

1. **Recupero di dati sui film e programmi televisivi**: Le API consentono agli sviluppatori di accedere a dettagli come titoli, descrizioni, immagini, date di rilascio, generi, punteggi degli utenti e molto altro ancora per film e programmi televisivi presenti nel database di TMDb.

2. **Recupero di informazioni sulle persone**: Gli sviluppatori possono ottenere dettagli su attori, registi, sceneggiatori e altre persone coinvolte nella produzione cinematografica.

3. **Recupero di immagini e poster**: Le API consentono di accedere a immagini, poster e manifesti relativi ai film e ai programmi televisivi nel database di TMDb, utile per la creazione di gallerie o per l'implementazione di funzionalità di visualizzazione delle immagini.

4. **Ricerca avanzata**: Le API supportano funzionalità di ricerca avanzata, consentendo agli sviluppatori di cercare film, programmi televisivi o persone utilizzando diversi criteri come titoli, parole chiave, anno di rilascio e altro ancora.

5. **Recupero delle tendenze e dei film popolari**: Gli sviluppatori possono accedere alle informazioni sui film più popolari, le tendenze attuali e i film in arrivo.

6. **Gestione degli account utente**: Le API consentono agli sviluppatori di integrare funzionalità di account utente, permettendo agli utenti di autenticarsi, aggiungere film alla lista dei preferiti, votare e altro ancora.

Le API di TMDb utilizzano protocolli standard come HTTP e JSON per le richieste e le risposte, rendendo l'integrazione con qualsiasi linguaggio di programmazione relativamente semplice.

**Per utilizzare al meglio le API di TMDb, è consigliabile consultare la documentazione ufficiale.** [(click here)](https://developer.themoviedb.org/docs/getting-started)

#### Esempio: Ricerca per nome

```python
import requests

def get_movie_details(movie_name):
    api_key = "6418576c12ada8e0deaed2e2eb6f634c"
    base_url = "https://api.themoviedb.org/3"
    
    # Effettua una ricerca per il nome del film
    search_url = f"{base_url}/search/movie"
    params = {
        "api_key": api_key,
        "query": movie_name
    }
    response = requests.get(search_url, params=params)
    data = response.json()

    if data['results']:
        # Prendi il primo risultato della ricerca
        movie_id = data['results'][0]['id']

        # Ottieni i dettagli del film
        movie_details_url = f"{base_url}/movie/{movie_id}"
        params = {
            "api_key": api_key
        }
        response = requests.get(movie_details_url, params=params)
        movie_details = response.json()

        # Stampa le caratteristiche del film
        print("Titolo:", movie_details['title'])
        print("Data di uscita:", movie_details['release_date'])
        print("Voto medio:", movie_details['vote_average'])
        print("Trama:", movie_details['overview'])

        # Ottieni e stampa la locandina del film
        poster_path = movie_details['poster_path']
        if poster_path:
            poster_url = f"https://image.tmdb.org/t/p/w500/{poster_path}"
            print("Locandina:", poster_url)
        else:
            print("Locandina non disponibile")
    else:
        print("Film non trovato")

if __name__ == "__main__":
    movie_name = "Quo Vado"
    get_movie_details(movie_name)
```

#### Esempio: Recensioni per un determinato film

```python
import requests

def get_movie_reviews(movie_id):
    api_key = "6418576c12ada8e0deaed2e2eb6f634c"
    base_url = "https://api.themoviedb.org/3"
    
    # Ottieni le recensioni del film specificato
    reviews_url = f"{base_url}/movie/{movie_id}/reviews"
    params = {
        "api_key": api_key
    }
    response = requests.get(reviews_url, params=params)
    data = response.json()

    if 'results' in data:
        # Stampa le recensioni
        if data['results']:
            print("Recensioni per il film:")
            for review in data['results']:
                print("Autore:", review['author'])
                print("Contenuto:", review['content'])
                print("----------------------")
        else:
            print("Nessuna recensione disponibile per questo film.")
    else:
        print("Recensioni non trovate")

if __name__ == "__main__":
    # Specifica l'ID del film desiderato
    movie_id = 550  # ID di "Fight Club" come esempio
    get_movie_reviews(movie_id)
```

#### 3.1 Wrapper, cosa sono e come possono aiutarci.

Un wrapper API offre solitamente una serie di funzioni predefinite che incapsulano le chiamate HTTP e semplificano l'accesso ai dati forniti dalle API. Inoltre, i wrapper possono includere logica aggiuntiva per gestire errori, autenticazione, paginazione e altre operazioni comuni.
L'obiettivo principale di un wrapper API è quello di semplificare e rendere più intuitivo l'utilizzo delle API, consentendo agli sviluppatori di concentrarsi sulla logica del proprio software piuttosto che sul dettaglio delle chiamate HTTP e sul parsing dei dati di risposta.

#### Analisi Budget vs Incassi

```python
import json
import matplotlib.pyplot as plt
from tmdbv3api import TMDb, Movie
import requests

tmdb = TMDb()
tmdb.api_key = "6418576c12ada8e0deaed2e2eb6f634c"
movie = Movie()

movie_ids = [550, 680, 155, 13, 27205]  # Esempi di ID di film
movies_data = {}

# Ottenere i dati per ogni film
for movie_id in movie_ids:
    try:
        # Ottieni i dettagli del film
        movie_details = movie.details(movie_id)
        
        # Ottieni il budget e il revenue del film
        budget = movie_details.budget
        revenue = movie_details.revenue
        
        # Aggiungi i dati del film al dizionario
        movies_data[movie_details.title] = {
            "Budget": budget,
            "Revenue": revenue
        }
    except Exception as e:
        print(f"Errore nell'ottenere i dettagli del film con ID {movie_id}: {e}")


# Creare i dati per il grafico
film_titles = list(movies_data.keys())
budgets = [movies_data[film]["Budget"] for film in film_titles]
revenues = [movies_data[film]["Revenue"] for film in film_titles]
x = range(len(film_titles))

# Grafico a barre affiancate
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(x, budgets, color='b', width=0.4, label='Budget')
plt.bar([i + 0.4 for i in x], revenues, color='r', width=0.4, label='Revenue')
plt.xlabel('Film')
plt.ylabel('Valore')
plt.title('Budget vs Incasso (Barre affiancate)')
plt.xticks([i + 0.2 for i in x], film_titles, rotation=45, ha='right')
plt.legend()

# Grafico a scatter plot
plt.subplot(1, 2, 2)
plt.scatter(budgets, revenues, color='g')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Budget vs Incasso (Scatter plot)')

for i, film_title in enumerate(film_titles):
    plt.text(budgets[i], revenues[i], film_title, fontsize=9, ha='left', va='bottom')

# Mostrare i grafici
plt.tight_layout()
plt.show()
```

#### Analisi Punteggio medio per genere

```python
import matplotlib.pyplot as plt
from tmdbv3api import TMDb, Movie

tmdb = TMDb()
tmdb.api_key = "6418576c12ada8e0deaed2e2eb6f634c"

# Inizializza l'oggetto Movie
movie = Movie()

generi_ids = {
    "Azione": 28,
    "Commedia": 35,
    "Drammatico": 18,
    "Fantascienza": 878,
    "Horror": 27
}
punteggi_per_genere = {}
popular_movies = movie.popular()

# Ottieni il punteggio di valutazione medio per ogni genere
for genere, genere_id in generi_ids.items():
    try:
        # Filtra i film per genere
        film_per_genere = [film for film in popular_movies if genere_id in film.genre_ids]
        
        # Calcola il punteggio di valutazione medio per il genere
        punteggio_medio = sum([film.vote_average for film in film_per_genere]) / len(film_per_genere)
        
        # Aggiungi il punteggio di valutazione medio al dizionario
        punteggi_per_genere[genere] = punteggio_medio
    except Exception as e:
        print(f"Errore nell'ottenere i film per il genere {genere}: {e}")

# Creare il grafico a barre
plt.figure(figsize=(10, 6))
plt.bar(punteggi_per_genere.keys(), punteggi_per_genere.values(), color='skyblue')
plt.xlabel('Genere')
plt.ylabel('Punteggio Medio')
plt.title('Confronto Punteggi Medi per Genere di Film')
plt.ylim(0, 10)  # Imposta il range dell'asse y da 0 a 10 per i punteggi
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Aggiungi una griglia sull'asse y
plt.tight_layout()

# Mostrare il grafico
plt.show()
```

### Analisi popolarità delle lingue

```python
import matplotlib.pyplot as plt
import requests

url = "https://api.themoviedb.org/3/movie/popular?language=en-US&page=10"

headers = {
    "accept": "application/json",
    "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiI2NDE4NTc2YzEyYWRhOGUwZGVhZWQyZTJlYjZmNjM0YyIsInN1YiI6IjY1ZmM0YzE1NzcwNzAwMDE3YzA4OGU4OSIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.tApgs5aDyar60IpfyJIH_T3FEIuJ4NCghvCNUoPgodM"
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    data = response.json()

    # Estrai le lingue dei film
    languages = {}
    for movie in data['results']:
        original_language = movie['original_language']
        if original_language in languages:
            languages[original_language] += 1
        else:
            languages[original_language] = 1

    # Visualizza il grafico a torta delle lingue dei film
    labels = languages.keys()
    counts = languages.values()

    plt.figure(figsize=(10, 6))
    plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')
    plt.title('Distribuzione delle lingue tra i film più popolari')
    plt.show()
else:
    print("Errore nella richiesta:", response.status_code)
```
